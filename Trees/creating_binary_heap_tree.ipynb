```python
from collections import deque
class Queue:
    def __init__(self):
        self.q = deque()

    def enq(self,value):
        self.q.appendleft(value)

    def deq(self):
        if len(self.q) > 0:
            return self.q.pop()
        else:
            return None

    def __len__(self):
        return len(self.q)


class Node:
        
    def __init__(self,key = None,value = None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None
        
    def set_value(self,value):
        self.value = value
        
    def get_value(self):
        return self.value
        
    def set_left_child(self,left):
        self.left = left
        
    def set_right_child(self, right):
        self.right = right
        
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right

    def has_left_child(self):
        return self.left != None
    
    def has_right_child(self):
        return self.right != None
```

```python
class Minheap:
    
    #creating a new empty heap or heap with root node initialized
    def __init__(self,key=None,value=None):
        # self.arr = [None for _ in range(freq_len)]  #using an array to build a heap with size equal to frequency dictionary
        self.arr = []
        self.root = Node(key,value)
        self.num_nodes = 0

    ##common operations are:
    # Find minimum which is the root node (peek)
    #operation without modifying the heap
    def find_min(self):
        return self.root

    #insert a new key to the heap
    #NB: need to maintain heap property during insertion
    def push(self,frequency):
        """
        Accepts either a dictionary acting as a frequency container for various keys,
        or a Node object.
        """
        if type(frequency) == dict:
            for key,value in frequency.items(): 
                self.binary_heap_tree(key,value)
            self.bfs()
        else:
            if self.root == None or self.num_nodes == 0:
                self.root = frequency
                self.num_nodes += 1
            else:
                node = self.pop()
            self.num_nodes += 1
            #traversing through the tree
            while node:
                #ensuring heap property for min_heap is observed
                if node.value <= frequency.value:
                    if not node.has_left_child():
                        node.left = frequency
                        return
                    else:
                        node.right = frequency
                        return
                else:
                    parent = frequency
                    child = node
                    if child == self.root:
                        self.root = parent
                    if not parent.has_left_child():
                        parent.left = child
                        return
                    else:
                        parent.right = child
                        return
            
            self.bfs()

    #making binary treee from string message
    def binary_heap_tree(self,key,value):
        new_node = Node(key,value)
        #insert premier node
        if self.root == None or self.num_nodes == 0:
            self.root = new_node
            self.num_nodes += 1
        else:
            node = self.root
            self.num_nodes += 1
            #traversing through the tree
            while node:
                #ensuring heap property for min_heap is observed
                if node.value <= new_node.value:
                    if not node.has_left_child():
                        node.left = new_node
                        return
                    else:
                        node.right = new_node
                        return
                else:
                    parent = new_node
                    child = node
                    if child == self.root:
                        self.root = parent
                    if not parent.has_left_child():
                        parent.left = child
                        return
                    else:
                        parent.right = child
                        return
    
    #breadth first search of binary_heap_tree to return an array
    def bfs(self):
        q = Queue()
        # visit_order = list()
        q.enq(self.root)

        while len(q) != 0:
            node = q.deq()
            # visit_order.append(node.value)
            self.arr.append(node)
            if node.has_left_child():
                q.enq(node.get_left_child())
            if node.has_right_child():
                q.enq(node.get_right_child())
```
